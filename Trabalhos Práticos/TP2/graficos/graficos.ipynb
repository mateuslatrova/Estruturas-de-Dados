```python
import numpy as np
import matplotlib.pyplot as plt

fig1 = plt.figure()

data = [0.0003, 0.007, 0.0095, 0.177, 0.674, 16.85, 97, 427]

x = ['100', '500','1000', '5000', '10000', '50000','100000', '200000']
xi = list(range(len(x)))

X = np.arange(8)

ax = fig1.add_axes([0,0,1,1])

ax.bar(X + 0.00, data, color = 'm', width = 0.25)
ax.set_ylabel('Tempo de execução(s)')
ax.set_xlabel('Número de elementos - N')
ax.set_title('Insertion Sort - Dados em ordem crescente')
plt.xticks(xi, x)

plt.savefig('insertion1crescente.png', format='png')
plt.show()
```

```python
#insertion sort - all 3 files together:
import matplotlib.transforms as mtrans

data = [[0.0003, 0.007, 0.0095, 0.177, 0.674, 16.85, 97, 427], 
        [0, 0.000393, 0.0029, 0.040, 0.111, 1.7, 7.892, 33.081],
        [0.000002, 0.0044, 0.017, 0.127, 0.478, 11.293, 52.164, 223.428]]

z = ['100', '500','1000', '5000', '10000', '50000','100000', '200000']
z_len = list(range(len(z))) 

X = np.arange(8)
fig4 = plt.figure()
ax = fig4.add_axes([0,0,1,1])

ax.bar(X + 0.333, data[0], color = 'm', width = 0.25)
ax.bar(X + 0.666, data[1], color = 'c', width = 0.25)
ax.bar(X + 1, data[2], color = 'b', width = 0.25)

ax.set_ylabel('Tempo de execução(s)')
ax.set_xlabel('Número de elementos - N')
ax.set_title('InsertionSort - Comparação dos 3 casos')

ax.legend(labels=['Ordem Crescente', 'Ordem Descrescente', 'Ordem Aleatória'])
plt.xticks(z_len, z)

trans = mtrans.Affine2D().translate(27, 0)
for t in ax.get_xticklabels():
    t.set_transform(t.get_transform()+trans)

plt.savefig('insertionComp.png', format='png')
plt.show()
```
